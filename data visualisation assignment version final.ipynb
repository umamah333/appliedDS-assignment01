```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("heart.csv")

#line graph representing the relation of cholestrol with thalassemia patient
    
def customlinegraph(ch,th):
    '''this is a line graph function which is used to show the relation of cholestrol and thalassemia in heart disease patients'''
    data.chol.plot(kind = 'line', color = 'g',label = 'chol',linewidth=1,alpha = 0.7,grid = True,linestyle = ':')
    data.thalach.plot(color = 'r',label = 'thalach',linewidth=1, alpha = 0.7,grid = True,linestyle = '-.')
    plt.legend(loc='upper right')
    plt.title('Relation of Cholestrol with Thalach', fontsize = 20)
    plt.savefig('linegraph.jpg')
    plt.show()
    
    
# Scatter Plot
#relation of cholestrol with age
# x = Age, y = Cholesterol
def customplot1(x,y):
    '''this the custom made function which print the scatter plot for the cholestrol level with respect to age.'''
    data.plot(kind = 'scatter',x = 'age', y = 'chol',alpha = 0.5, color = 'red')
    plt.xlabel('Age')
    plt.ylabel('Cholesterol')
    plt.title('Age-Cholesterol Plot')
    plt.savefig('scatterplot.jpg')
    plt.show()
    
    

#histogram
#age frequency
#frequency of aged people having heart disease


def customhisto(y):
    '''this histogram function is used to visualise the frequency of heart disease within different age group people'''
    data.age.plot(kind = "hist",bins = 50, figsize = (12,12))
    plt.xlabel('frequency with respect to age')
    plt.ylabel('age')
    plt.title('frequency of heart disease with respect to age')
    plt.savefig('histogram.jpg')
    plt.show()
    
    
print(data)
```
