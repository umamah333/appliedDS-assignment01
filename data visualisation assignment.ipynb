```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("heart.csv")
print(data)
```

```python
data.head()#displays top 5 rows of the dataset
data.shape
```

```python
data.isnull().sum() #shows null values in dataset which are o means no null values are there in the dataset
```

```python
data_dupval = data.duplicated().any() #displays any duplicated values in the datatset
print(data_dupval)
data = data.drop_duplicates()#removes duplicated values from the dataset
data.shape
```

```python
#line graph representing the relation of cholestrol with thalassemia patients
data.thal.plot(kind = 'line', color = 'r', label = ' Thalassemia',linewidth = 1, alpha = 0.5, grid = True, linestyle = '--')
data.oldpeak.plot(color = 'b',label = 'Cholesterol',linewidth = 1, alpha = 0.5, grid = True, linestyle = ':')
plt.legend(loc='upper left')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Thalassemia-Cholesterol Plot')
plt.show()
```

```python
# Scatter Plot
#relation of cholestrol with age
# x = Age, y = Cholesterol

data.plot(kind = 'scatter',x = 'age', y = 'chol',alpha = 0.5, color = 'red')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('Age-Cholesterol Plot')
plt.show()
```

```python
#histogram
#age frequency
#frequency of aged people having heart disease


data.age.plot(kind = "hist",bins = 50, figsize = (12,12))
plt.show()
```
